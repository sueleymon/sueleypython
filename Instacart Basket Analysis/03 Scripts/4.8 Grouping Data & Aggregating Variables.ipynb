## 03. Grouping data +  Aggregating Data with agg() & transform() + Deriving Columns with loc

```python
# Show the current data frame
ords_prods_merge.head()
```

```python
# Group the deparment_id by the mean order_amount of the whole database
ords_prods_merge.groupby('department_id').agg({'order_amount': ['mean']})
```

## 3) The difference are that more departments show up and the means are different from the subset, because the database his way bigger than the small subset we created and therefore doesn't show the true values of the database.

```python
# Create a max_order column and group it by user_id and the order_amount
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_amount'].transform(np.max)
```

```python
# Check if it worked
ords_prods_merge.head(15)
```

```python
# Create a new column with conditions for loyal, regular and new customers through the max_order value column
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check data consistency 
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check if creating the column worked
ords_prods_merge.head()
```

```python
# Find the basic statistics from the values of the loyalty_flag
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Create a new column for average orders and group it by user_id and the mean of prices
ords_prods_merge['avg_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create a new column through the avg order column to determine the spending_flag column and their conditions
ords_prods_merge.loc[ords_prods_merge['avg_order'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check the data for consistency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check if the new columns worked
ords_prods_merge.head(50)
```

```python
# Create a new column for median days and group it by user_id and the median of the days since prior order column
ords_prods_merge['med_days'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Create a new column through the med days column to determine the order_frequency_flag column and their conditions
ords_prods_merge.loc[ords_prods_merge['med_days'] > 20, 'order_frequency_flag'] = 'Non-frequent customer.'
```

```python
ords_prods_merge.loc[(ords_prods_merge['med_days'] <= 20) & (ords_prods_merge['med_days'] > 10), 'order_frequency_flag'] = 'Regular customer.'
```

```python
ords_prods_merge.loc[ords_prods_merge['med_days'] <= 10, 'order_frequency_flag'] = 'Frequent customer.'
```

```python
# Check for data consistency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check if the new columns worked
ords_prods_merge.head(50)
```
