```python
# Rename the columns so they are more logical
customer.rename(columns = {'First Name' : 'first_name'}, inplace = True)
customer.rename(columns = {'Surnam' : 'surname'}, inplace = True)
customer.rename(columns = {'Gender' : 'gender'}, inplace = True)
customer.rename(columns = {'Age' : 'age'}, inplace = True)
customer.rename(columns = {'STATE' : 'state'}, inplace = True)
```

```python
# Check for data consistency and if it worked
customer.head()
```

```python
# Use the function from Exercise 4.5 to check for mixed data
for col in customer.columns.tolist():
  weird = (customer[[col]].applymap(type) != customer[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customer[weird]) > 0:
    print (col)
```

```python
# Check if there are null values
customer.isnull().sum()
```

```python
# Create a subset that shows the null values in the column 'first_name'
df_nan = customer[customer['first_name'].isnull() == True]
```

```python
df_nan
```

```python
# Impute NaN value with 'Not filled'
customer['first_name'].fillna('Not filled', inplace=True)
```

```python
# The impute of values worked and none are 
customer.isnull().sum()
```

```python
# Check for duplicates
customer[customer.duplicated()]
```
