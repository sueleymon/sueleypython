## 03. Create bar charts, histogram, scatterplots and line chart

```python
# Create a histogram that covers sales
hist = ords_prods_cust_merge['order_hour_of_day'].plot.hist(bins = 24) 
```

### This histogram shows information about the x axis "order_hour_of_day" and the y axis frequency. This means this is a representation of at which time a sale was made. We see in the early hours not having many sales but the frequency starts to increase around 7-8 am and it reaches it's peak hours from 10-17 while dropping afterwards.

```python
# Save the histogram as a png file
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'histogram_order_hour_of_day.png'))
```

```python
# Create a bar chart for the 'loyalty_flag' column
bar = ords_prods_cust_merge['loyalty_flag'].value_counts().plot.bar(color =['red', 'blue', 'green'])
```

### The bar chart shows that regular customers purchase the most, followed by loyal customer and new customer purchase the least

```python
# Save the bar chart as a png file
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_chart_loyalty_flag.png'))
```

```python
# Set the random seed as 4, so it's able to be reproduced by other members
np.random.seed(4)
```

```python
# Set a variable that splits the data in a 70/30 Split
dev = np.random.rand(len(ords_prods_cust_merge)) <= 0.7
```

```python
# Set 70% of the data as the variable big
big = ords_prods_cust_merge[dev]
```

```python
# Set 30% of the data as the variable small
small = ords_prods_cust_merge[~dev]
```

```python
# Return the number of items/rows inside this data frame
len(ords_prods_cust_merge)
```

```python
# Return the number of items/rows inside these combined data frames to check if it worked
len(big) + len(small)
```

```python
# Create a subset of the small data with the columns 'orders_hour_of_day' and 'prices' only
df = small[['order_hour_of_day', 'prices']]
```

```python
line = sns.lineplot(data = df, x = 'order_hour_of_day', y = 'prices')
```

### There is some correlation showing that the prices are the most volatile in the early mornings and have a more consistent pattern later on. But you can't really tell if there is a significant increase because the highest difference would be 10 cents and that's not enough to justify that train of thought.

```python
# Save the line chart as a png file
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_chart_order_hour_of_day_prices.png'))
```

```python
# Create a subset of the small data with the columns 'n_dependants' and 'age' only
df_2 = small[['age' , 'n_dependants']]
```

```python
line_2 = sns.lineplot(data = df_2, x = 'age',y = 'n_dependants')
```

### There is no connection between the number of dependants and age

```python
# Save the line chart as a png file
line_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_chart_number_of_dependents_age.png'))
```

```python
# Create a scatterplot with the columns 'age' and 'income' only
scatterplot = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_cust_merge)
```

### After the age of 40 the income cap of 400000 gets surpassed for the first time. The income after 40 is more increased compared to the age below 40.

```python
# Save the scatterplot as a png file
scatterplot.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatterplot_age_income.png'))
```
