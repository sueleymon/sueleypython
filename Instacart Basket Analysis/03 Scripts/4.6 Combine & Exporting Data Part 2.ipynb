```python
# Import the pickle file
df_ords = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = 0)
```

## 03. Combine data sets

```python
df_ords.shape
```

```python
df_prods.shape
```

```python
df_ords.head()
```

```python
df_prods.head()
```

```python
# Merge the two data sets together
df_merged_large = df_ords.merge(df_prods, on = 'product_id', indicator = 'merge_flag')
```

```python
df_merged_large
```

```python
# Check the merge for errors
df_merged_large['merge_flag'].value_counts()
```

```python
# Create a new data frame without the merge columns 
df_merged_large_without_merge_column = df_merged_large.drop(columns = ['_merge', 'merge_flag'])
```

```python
df_merged_large_without_merge_column.head()
```

```python
# Export data to pkl

df_merged_large_without_merge_column.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
