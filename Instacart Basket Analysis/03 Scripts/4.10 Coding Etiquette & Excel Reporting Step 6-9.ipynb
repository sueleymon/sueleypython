### 06. Compare your customer profiles with regions and departments. Can you identify a link between them? Customers may demonstrate different behaviors across regions, which could make this information important(Step 8 + 9)

```python
# Create a crosstab to visualize the table with the columns 'customer_profile' and 'region'
crosstab_1 = pd.crosstab(final['customer_profile'], final['region'], dropna = False)
```

```python
crosstab_1
```

### There is too much information to distinquish a lot here, but most customer are from middle-income and south/west

```python
# Visualize the results of the crosstab
customer_region_bar = crosstab_1.plot.bar(stacked = True)
plt.title("Customer Profile by Region")
plt.xlabel("Customer Profile")
```
