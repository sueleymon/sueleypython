## 03. Create subsets and work with new functions

```python
# Subset of data so it's faster to analyze
df = orders_products_merged[:1000000]
```

```python
# Defining a new function which categorize items in different price ranges
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Use the function and create a new column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df.shape
```

```python
# Count the price ranges
df['price_range'].value_counts(dropna = False)
```

```python
# Check the highest price on why there were no high-range products
df['prices'].max()
```

```python
# Use the loc function to define our range for our subset of data
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Check if it worked
df['price_range_loc'].value_counts(dropna = False)
```
